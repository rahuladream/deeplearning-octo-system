## System of Equations

The system of linear equations we'll solve can be represented like:

ax+by=c
 
dx+ey=f
 
where  a,b,c,d,e, and f  are real valued constants, and neither  a and b  nor  d and e  are both  0 .

Now, suppose we want to solve for  x and y  where:

−x+3y=7
 
3x+2y=1

Working out some algebra, we have:

y=13x+73
 
and

y=−32x+12

```python
import numpy as np
np.random.seed(0)
```

```python
xs = np.linspace(2, 2.5, 100)
```

```python
xs
```

```python
y1 = [((1/3) * x + 7/3) for x in xs]
```

```python
y2 = [(-1*(3/2) * x + 1/2) for x in xs]
```

```python
y1
```

```python
print("Some points from the first line: ")
print([i for i in zip(xs[:3], y1[:3])])

print("\n Some points from the second line: ")
print([i for i in zip(xs[:3], y2[:3])])
```
